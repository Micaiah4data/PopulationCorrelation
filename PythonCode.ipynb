```python
data.sort_values(by=['growthRate'], inplace = False, ascending = False)
```

```python
plt.scatter(x=data['growthRate'], y=data['pop2023'])

plt.title('Growth Rate vs 2023 Population')
plt.xlabel('2023 Population')
plt.ylabel('Growth Rate')
plt.show()
```

```python
sns.regplot(x="growthRate", y="pop2023", data=data, scatter_kws={"color": "purple"}, line_kws={"color":"Black"})
```

```python
data.corr(method='spearman') #pearson, kendall, spearman
```

```python
correlation_matrix = data.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for the Data')
plt.xlabel('Countries Population')
plt.ylabel('Countries Populatio')

plt.show()
```
